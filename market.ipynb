```python
import csv
```

```python
def read_data(path):
    """
    Membaca data dari csv file

    parameter:
        path (str): lokasi file
    
    result:
        list_data (list): hasil data dari file dalam bentuk list
    """
    list_data = []
    with open(path, 'r') as csvfile:
        csvreader = csv.DictReader(csvfile)

        for row in csvreader:
            list_data.append(row)

    return list_data
```

```python
data = read_data("Mall_Customers.csv")
data
```

```python
def drop_data(data):
    """
        membuang kolom customerID, Gender, Age

        paremeter:
            data (list): list data

        result:
            annual_income (list): list data annual income
            spending_score (list): list data spending score
    """
    annual_income = []
    spending_score = []
    for num in range(0 , len(data)):
        annual_income.append(int(float(data[num]["Annual Income (k$)"])))
        spending_score.append(int(float(data[num]["Spending Score (1-100)"])))
        
    return annual_income, spending_score


    
```

```python
annual_income, spending_score = drop_data(data)
```

```python
def karakter_target(annual_income, spending_score):
    """
    Mencari target dengan annual_income dan spending_score

    parameter:
        annual_income (list): list daftar annual income
        spendin_scorre (list): list daftar spending score

    return:
        target (list): banyaknya orang yang sesuai target [annual_income, spending_score]
    """
    target = []
    for idx_annual, val_annual in enumerate(annual_income):
        for idx_spending, val_spending in enumerate(spending_score):
            if idx_annual == idx_spending:
                target.append([val_annual, val_spending])
                    

    return target
```

```python
#ambil target berdasarkan annual_income dan spending score
target = karakter_target(annual_income, spending_score)
```
